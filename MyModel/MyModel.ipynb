# Plotting B-scan

Terdapat folders ``tools`` dengan modul ``plot_Bscan`` yang dapat digunakan untuk memplotting B-scan model. 
Plottingan B-scan dapat dilakukan dengan menjalankan command line di bawah ini : 

    python -m tools.plot_Bscan user_models/cylinder_Bscan_2D_merged.out Ez
Hasil dari Plottingan dapat disimpan juga dalam bentuk PDF maupun PNG

```python
%matplotlib inline
import os
from tools.plot_Bscan import get_output_data, mpl_plot

filename = os.path.join('output', 'data100.out')
rxnumber = 1
rxcomponent = 'Ez'
outputdata, dt = get_output_data(filename, rxnumber, rxcomponent)
plt = mpl_plot(filename,outputdata, dt, rxnumber, rxcomponent)

# Change from the default 'seismic' colormap
plt.set_cmap('gray')

(path, filename) = os.path.split(filename)
savefile = os.path.splitext(filename)[0]
plt.savefig(path + os.sep + savefile + '.png', dpi=150, format='png', bbox_inches='tight', pad_inches=0.1)
```
